```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("happyscore_income.csv")

data.head()
```

```python

gdp = data["GDP"]
hscore = data["happyScore"]
plt.plot(gdp,hscore,".")
plt.xlabel("GDP")
plt.ylabel("Happiness Score")
plt.show()
```

```python

income = data["avg_income"]
hscore = data["happyScore"]
plt.plot(income,hscore,".")
plt.xlabel("Income")
plt.ylabel("Happiness Score")
plt.show()
```

```python

income = data["avg_income"]
sat = data["avg_satisfaction"]
plt.plot(income,sat,".")
plt.xlabel("Income")
plt.ylabel("Satisfaction")
plt.show()
```
