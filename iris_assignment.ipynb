```python
import pandas as pd
import matplotlib.pyplot as plt

iris_data = pd.read_csv("Iris.csv")

iris_data.plot(kind = "scatter", x = "PetalLengthCm", y = "PetalWidthCm")

plt.grid()
```

```python
import numpy as np

a = ["Iris-setosa", "Iris-versicolor", "Iris-virginica"]

is_x, is_y, ive_x, ive_y, ivi_x, ivi_y = [], [], [], [], [], []

for e,i in enumerate(iris_data["Species"]):
    if(i == a[0]):
        is_x.append(iris_data["PetalLengthCm"][e])
        is_y.append(iris_data["PetalWidthCm"][e])
    elif(i == a[1]):
        ive_x.append(iris_data["PetalLengthCm"][e])
        ive_y.append(iris_data["PetalWidthCm"][e])
    elif(i == a[2]):
        ivi_x.append(iris_data["PetalLengthCm"][e])
        ivi_y.append(iris_data["PetalWidthCm"][e])

plt.plot(is_x, is_y, 'b*', label = "Iris-setosa")
plt.plot(ive_x, ive_y, 'g.', label = "Iris-versicolor")
plt.plot(ivi_x, ivi_y, 'rv', label = "Iris-virginica")

plt.legend()
```

```python
import statistics as st # -> for calculating statistical mode (not supported in numpy)

print(iris_data.head())


print("\n---------\n")

# calculating mean
for i in iris_data:
    if(type(iris_data[i][1]) == type(iris_data["SepalLengthCm"][1])):
        print("Mean of " ,i, " is ", np.mean(iris_data[i]))

print("\n---------\n")

# calculating median
for i in iris_data:
    if(type(iris_data[i][1]) == type(iris_data["SepalLengthCm"][1])):
        print("Median of " ,i, " is ", np.median(iris_data[i]))

print("\n---------\n")

# calculating mode
for i in iris_data:
    if(type(iris_data[i][1]) == type(iris_data["SepalLengthCm"][1])):
        print("Mode of " ,i, " is ", st.mode(iris_data[i]))

```
